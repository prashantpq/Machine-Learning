```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

iris = sns.load_dataset('iris')
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=iris)
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.title('Frequency of Each Class Label in Iris Data')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='sepal_width', y='petal_width', data=iris)
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.title('Petal Width vs Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=iris['petal_length'], shade=True)
plt.xlabel('Petal Length')
plt.ylabel('Density')
plt.title('Density Distribution of Petal Length')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.pairplot(data=iris, hue='species')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(iris.iloc[:, :-1].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()
```
